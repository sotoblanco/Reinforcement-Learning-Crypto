```python
# To fetch financial data
import yfinance as yf

# For visualisation
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn-darkgrid')

# Download the minute data for Apple
nq_minute_data = yf.download(tickers="NQ=F", period="5d", interval="1m", auto_adjust=True)

# Dispaly the first 5 rows
nq_minute_data.head()
```

```python
# Aggregate function
ohlcv_dict = {'Open': 'first',
              'High': 'max',
              'Low': 'min',
              'Close': 'last',
              'Volume': 'sum'
             }
```

```python
# Resample data to 15 minutes data
nq_minute_data_15M = nq_minute_data.resample('15Min').apply(ohlcv_dict).dropna()
nq_minute_data_15M.head()
```

```python
nq_minute_data_1H = nq_minute_data.resample('1H').apply(ohlcv_dict).dropna()
nq_minute_data_1H.head()
```

```python
# Resample data to 4 hours data
nq_minute_data_4H = nq_minute_data.resample('4H').apply(ohlcv_dict).dopna()

# Display the first 5 rows
nq_minute_data_4H.head()
```
